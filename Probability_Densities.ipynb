#**Probability Density Function** (PDF)

If we want to obtain the probability for the variable appearing in a certain interval, we have to integrate the PDF over that range.

e.g. What is the probability that a girl is between 40 and 45 kilograms


```
#import the modules we may use
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import math
```

```
#Creating normal distribution with the mean of 47 and standard deviation of 3
norm_dist = stats.norm(47, 3)
x = np.linspace(38, 56, 100)
y = norm_dist.pdf(x)

#Plotting the PDF of the given normal distribution
sns.set()
plt.figure(figsize = (12, 4))
plt.plot(x, y)

#Emphasizing the area
plt.axvline(x = 40, color = "red")
plt.axvline(x = 45, color = "red")

plt.show()

#Calculating the designated area
upper = norm_dist.cdf(45)
lower = norm_dist.cdf(40)

print("The probability that a girl is between 40 and 45 kilograms is", round(upper - lower, 3) * 100, "%")
```

# **Cumulative Distribution Function** (CDF)

which gives the probability of obtaining a values smaller than the given value.

e.g. What is the probability that a girl is less than 45 kilograms?

```
#承接上面設定的常態分佈
y2 = norm_dist.cdf(x)

#Plotting the CDF of the given normal distribution
plt.plot(x, y2)

#Emphasizing line
plt.axhline(y = norm_dist.cdf(45), color ="red")
plt.show()

print("The probability that a girl is less than 45 kilograms is", round(norm_dist.cdf(45), 3) * 100, "%")
```

# **Survival Function** (SF) = 1 - CDF

which gives the probability of obtaining a value larger than the given value.

e.g. What is the probability that a girl is higher than 45 kilograms?

```
#承接上面設定的常態分佈
y3 = norm_dist.sf(x)

#Plotting the SF of the given normal distribution
plt.plot(x, y3)

#Emphasizing line
plt.axhline(y = norm_dist.sf(45), color = "red")
plt.show()

print("The probability that a girl is higher than 45 kilograms is", round(norm_dist.sf(45), 3) * 100, "%")
```

# **Percentile Point Function**(PPF)

which is the inverse of the CDF. It can be used to answer the question like "Given a certain probability, what is the corresponding input value for the CDF?"

e.g. Given that I am looking for a girl who is lighter than 95% of all the other girl, what size does the subject have to be?

```
#承接上面設定的常態分佈
y2 = norm_dist.cdf(x)
y4 = norm_dist.ppf(y2)

#Plotting the PPF of the given normal distribution
plt.plot(y2, y4)

#Emphasizing line
plt.axvline(x = 0.05, color = "red")
plt.show()

print("The size that is that the girl who is lighter than 95% of all the other girl should be", round(norm_dist.ppf(0.05), 3))
```

# **Inverse Survival Function** (ISF)

e.g. Given that I am looking for a girl who is heavier than 95% of all the other girl, what size does the subject have to be?

```
#承接上面設定的常態分佈
y3 = norm_dist.sf(x)
y5 = norm_dist.isf(y3)

#Plotting the ISF of the given normal distribution
plt.plot(y3, y5)

#Emphasizing line
plt.axvline(x = 0.05, color = "red")
plt.show()

print("The size that is that the girl who is heavier than 95% of all the other girl should be", round(norm_dist.isf(0.05), 3))
```
